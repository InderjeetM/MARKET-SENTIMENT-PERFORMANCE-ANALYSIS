```python
import pandas as pd

print("🔄 Step 1: Converting dates to standard formats...")

# 1. Clean Fear & Greed Date
df_sentiment['clean_date'] = pd.to_datetime(df_sentiment['date']).dt.date

# 2. Clean Historical Trades Date (extracting just the day, ignoring time)
# Using dayfirst=True because '02-12-2024' represents December 2nd
df_historical['clean_date'] = pd.to_datetime(df_historical['Timestamp IST'], dayfirst=True).dt.date

print("🔗 Step 2: Merging your Trading Logs with Market Sentiment...")

# 3. Merge dataframes on the newly standardized 'clean_date' column
# We keep all historical trades ('left' join) and bring in matching sentiment data
df_merged = pd.merge(df_historical, df_sentiment[['clean_date', 'value', 'classification']], on='clean_date', how='left')

# Rename the columns slightly for clearer tracking
df_merged.rename(columns={'value': 'Fear_Greed_Score', 'classification': 'Market_Sentiment'}, inplace=True)

print(f"📊 Merge complete! Combined dataset size: {df_merged.shape[0]} rows")
print("\n👀 Preview of your complete, ready-to-analyze dataset:")
display(df_merged[['Timestamp IST', 'Coin', 'Side', 'Closed PnL', 'Fear_Greed_Score', 'Market_Sentiment']].head(5))
```

```python
import numpy as np
import pandas as pd

print("⚙️ Processing Metrics across Market Sentiments...")

# 1. Define a winning trade (where Closed PnL is strictly positive)
df_merged['Is_Win'] = df_merged['Closed PnL'] > 0

# 2. Group data by sentiment category and calculate key metrics
performance_summary = df_merged.groupby('Market_Sentiment').agg(
    Total_Trades=('Closed PnL', 'count'),
    Total_PnL=('Closed PnL', 'sum'),
    Average_PnL=('Closed PnL', 'mean'),
    Win_Rate=('Is_Win', lambda x: np.round(x.mean() * 100, 2))
).reset_index()

# Sort the categories logically from worst fear to highest greed
sentiment_order = ['Extreme Fear', 'Fear', 'Neutral', 'Greed', 'Extreme Greed']
performance_summary['Market_Sentiment'] = pd.Categorical(performance_summary['Market_Sentiment'], categories=sentiment_order, ordered=True)
performance_summary = performance_summary.sort_values('Market_Sentiment').reset_index(drop=True)

print("\n" + "="*70)
print("🏆 PRO-LEVEL TRADING INSIGHTS SUMMARY table")
print("="*70)
display(performance_summary)

print("\n" + "="*70)
print("💡 KEY ANALYSIS TAKEAWAYS TO COPIED FOR YOUR REPORT:")
print("="*70)
for index, row in performance_summary.iterrows():
    print(f"• In '{row['Market_Sentiment']}' environments, traders executed {row['Total_Trades']:,} trades with a Win Rate of {row['Win_Rate']}% and generated an Avg PnL of ${row['Average_PnL']:.4f}.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Image

# 1. Initialize Project Dataset
data = {
    'Market_Sentiment': ['Extreme Fear', 'Fear', 'Neutral', 'Greed', 'Extreme Greed'],
    'Total_Trades': [21400, 61837, 37686, 50303, 39992],
    'Average_PnL': [34.537862, 54.290400, 34.307718, 42.743559, 67.892861],
    'Win_Rate': [37.06, 42.08, 39.70, 38.48, 46.49]
}
df = pd.DataFrame(data)
sns.set_theme(style="whitegrid")

# ==============================================================================
# HALF 1: LINE CHART (Performance & Profit Efficiency)
# ==============================================================================
fig, ax1 = plt.subplots(figsize=(10, 5))

color_win = '#1f77b4'
ax1.set_xlabel('Market Sentiment Spectrum (Low to High Risk)', fontsize=12, fontweight='bold', labelpad=12)
ax1.set_ylabel('Win Rate (%)', color=color_win, fontsize=12, fontweight='bold')
line1 = ax1.plot(df['Market_Sentiment'], df['Win_Rate'], color=color_win, marker='o', linewidth=2.5, markersize=8, label='Win Rate (%)')
ax1.tick_params(axis='y', labelcolor=color_win)
ax1.set_ylim(30, 50)

# Create secondary Y-axis for PnL
ax2 = ax1.twinx()
color_pnl = '#2ca02c'
ax2.set_ylabel('Average PnL ($)', color=color_pnl, fontsize=12, fontweight='bold')
line2 = ax2.plot(df['Market_Sentiment'], df['Average_PnL'], color=color_pnl, marker='s', linewidth=2.5, markersize=8, linestyle='--', label='Avg PnL ($)')
ax2.tick_params(axis='y', labelcolor=color_pnl)
ax2.set_ylim(25, 75)

# Value Labels
for i, txt in enumerate(df['Win_Rate']):
    ax1.annotate(f"{txt}%", (df['Market_Sentiment'][i], df['Win_Rate'][i]), textcoords="offset points", xytext=(0,10), ha='center', fontweight='bold', color=color_win)
for i, txt in enumerate(df['Average_PnL']):
    ax2.annotate(f"${txt:.2f}", (df['Market_Sentiment'][i], df['Average_PnL'][i]), textcoords="offset points", xytext=(0,-15), ha='center', fontweight='bold', color=color_pnl)

lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper left')

plt.title('PROJECT PART 1: Strategy Performance & Profit Efficiency\n(Win Rate vs. Average PnL Tracking)', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('project_part1_line_chart.png', dpi=300)
plt.close() # Prevents duplicate notebook rendering

# ==============================================================================
# HALF 2: PIE CHART (Volume & Allocations)
# ==============================================================================
fig, ax = plt.subplots(figsize=(7, 7))

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0']
explode = (0, 0.04, 0, 0, 0.04) # Pop out key actionable segments

wedges, texts, autotexts = ax.pie(
    df['Total_Trades'], explode=explode, labels=df['Market_Sentiment'], autopct='%1.1f%%',
    startangle=140, colors=colors,
    textprops=dict(color="black", fontsize=11, weight="bold"),
    wedgeprops=dict(edgecolor='white', linewidth=1.5)
)

plt.title('PROJECT PART 2: Trade Execution & Activity Allocation\n(Where Do Traders Place Their Bets?)', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('project_part2_pie_chart.png', dpi=300)
plt.close() # Prevents duplicate notebook rendering

# ==============================================================================
# DISPLAY BOTH CHARTS IN JUPYTER CELL OUTPUT
# ==============================================================================
print("Showing Part 1 of the Project Showcase...")
display(Image(filename='project_part1_line_chart.png'))

print("\n" + "="*80 + "\n")

print("Showing Part 2 of the Project Showcase...")
display(Image(filename='project_part2_pie_chart.png'))
```
